```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
sqft = np.random.randint(500, 3000, size=100) 
price = 50 * sqft + 50000 + np.random.randint(-10000, 10000, size=100) 
```

```python
data = pd.DataFrame({'sqft': sqft, 'price': price})
```

```python
data.head()
```

```python
data.shape
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr_model = LinearRegression()
```

```python
X = data.drop('price', axis =1)
```

```python
y = data['price']
```

```python
lr_model.fit(X,y)
```

```python
val_to_predit = [[1200]]
lr_model.predict(val_to_predit)
```
